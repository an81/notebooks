# 4. Crypto Assets: Market Characteristics

```python
import pandas as pd
from datetime import datetime
import numpy as np
import pandas as pd

#import plotly 
import plotly.plotly 
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot

init_notebook_mode(connected=True)
```

```python
# load all coins into csv file
df = pd.read_csv("coins_all.csv")
```

```python
colors = ['#1b5eb7', '#1bb7a4', '#1bb751', '#88b71b', '#b7a21b', 
              '#b7661b', '#b7371b', '#b71b5c', '#9c1bb7', '#302333',
              '#9879ad', '#7a45f7', '#44d9f7', '#439af7', '#f74343',
              '#775303', '#757703', '#f3f70e', '#f7860d', '#e0af70']

data = []
gr = df.groupby("coin")
i = 0
for name, group in gr:
    trace1 = go.Scatter(
        name = name + " - low_volume",
        x = group.vol_market_cap_ratio[group.volume_label==0],
        y = group.vols, 
        
        mode = 'markers',
        marker = dict( size=3, color=colors[i], opacity=0.8),
    )
    
    trace2 = go.Scatter(
        name = name + " - normal_volume",
        x = group.vol_market_cap_ratio[group.volume_label==1],
        y = group.vols, 
        
        mode = 'markers',
        marker = dict( size=5, color=colors[i], opacity=0.5),
    )
    
    trace3 = go.Scatter(
        name = name + " - high_volume",
        x = group.vol_market_cap_ratio[group.volume_label==2],
        y = group.vols, 
        
        mode = 'markers',
        marker = dict( size=7, color=colors[i], opacity=0.2),
    )

    data.append(trace1)
    data.append(trace2)
    data.append(trace3)
    i = i + 1
    
layout = go.Layout(
    title = "Cryptoassets", 
    xaxis = {'title': 'volume/market_capitalisation'},
    yaxis = {'title': 'volatility'},
     )    

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
colors = ['#1b5eb7', '#1bb7a4', '#1bb751', '#88b71b', '#b7a21b', 
              '#b7661b', '#b7371b', '#b71b5c', '#9c1bb7', '#302333',
              '#9879ad', '#7a45f7', '#44d9f7', '#439af7', '#f74343',
              '#775303', '#757703', '#f3f70e', '#f7860d', '#e0af70']

data = []
gr = df.groupby("coin")
i = 0
for name, group in gr:
    trace = go.Scatter3d(
        name = name,
        x = group.volume_label,
        y = group.price_move_label,
        z = group.vols, 
        
        mode = 'markers',
        marker = dict( size=3, color=colors[i], opacity=0.25),
    )

    data.append(trace)
    i = i + 1
    
layout = go.Layout(
    title = "Cryptoassets",
    scene = dict(
        xaxis = {'title': 'volume'}, 
        yaxis = {'title': 'price_movement'},
        zaxis = {'title': 'volatility'},
     )    
)

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
